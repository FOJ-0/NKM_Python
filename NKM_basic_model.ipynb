## 1. Introduction <a name="introduction"></a>

The New Keynesian Synthesis (NKS) mixes the methodology of Real Business Cycles (RBC) with nominal and real rigidities to characterise short run macroeconomic developments.

The basic setup: three equation model
1. The NK IS equation: relates the evolution of the output gap to the interest rate. 
2. The NK Phillips curve: relates inflation to the output gap.
3. Taylor Rule: describes the conduct of the central bank, how the nominal interest rate is set by the central bank following fluctuations in the output gap and in the inflation rate.

This model is based on agents‘ micro founded decision rules where consumers maximize their welfare subject to an intertemporal budget constraint and where firms maximize their profit, subject to nominal rigidities, characterising the imperfect adjustment of prices on the goods market.

In contrast with the classical model, when prices are sticky the equilibrium path of real variables cannot be determined independently of monetary policy. Monetary policy is non-neutral. 

### Households

Assume representative infinitely-lived household

$$\max_{C_t, N_t, B_t} E_0 \sum_{t=0}^{\infty} \beta^{t} U(C_t , N_t)$$
s.t. 
$$P_t C_t + Q_t B_t \leq B_{t-1} + W_t N_t + T_t$$

Where
- $\beta$: is the discount factor.
- $C_t$: is the level of consumtion. 
- $N_t$: hours of work.
- $B_t$: are bond bought in period t. 
- $Q_t$: bond price in period t. 
- $W_t$: nominal wage.
- $T_t$: is a lump-sum component of income. 

Solvency condition: $\lim_{T\rightarrow\infty} E_t{B_T} \geq 0 $ for all t. 

#### FOC
- $C_t$: $\beta^t U_{ct} - \lambda_t P_t =0$
- $N_t$: $\beta^t U_{Nt} + \lambda_t W_t =0$
- $B_t$: $\lambda_t Q_t + E_t \lambda_{t+1} =0$

#### Optimal consumption and labor supply
The following equations determine the intertemporal consumtion allocation (Euler equation) and the labor-leisure choice. Together, tose equations determine the rational forward-looking household's allocation decisions. 
$$-\frac{U_{N,t}}{U_{c,t}}=\frac{W_t}{P_t}$$
$$Q_t=\beta E_t \left\lbrace \frac{U_{c,t+1}}{U_{c,t}}\frac{P_t}{P_{t+1}} \right\rbrace $$

Under the assumption of a period utility given by: 
$$ U(C_t, N_t)= \frac{C_t^{1-\sigma}}{1-\sigma} \frac{N_t^{1+\varphi}}{1+\varphi}$$

The log-linear version of the above optimality condition take the form: 
$$w_t-p_t = \sigma c_t + \varphi n_t$$
$$c_t = E_t c_{t+1} - \frac{1}{\sigma}(i_t-E_t(\pi_{t+1})-\rho)$$
Where
- $\rho \equiv -ln\beta$
- $i_t \equiv -lnQ_t$
- $\triangle c_{t+1} \equiv c_{t+1} - c_{t} \equiv lnC_{t+1}- lnC_{t}$
- $\pi_{t+1} \equiv p_{t+1} - p_t \equiv lnP_{t+1} - lnP_t$

#### Dynamic IS curve (DIS)

Combine the goods market clearing condition ($Y_t=C_t$) with the consumer's euler equation and $r_t\equiv i_t - E_t \pi_{t+1}$

$$y_t = E_t(y_{t+1}) - \frac{1}{\sigma}(i_t-E_t(\pi_{t+1})-\rho)$$
$$y^n_t = E_t(y^n_{t+1}) - \frac{1}{\sigma}(r_t^n-\rho)$$



$\tilde{y_t}\equiv y_t - y_t^n = [E_t(y_{t+1}) - \frac{1}{\sigma}(i_t-E_t(\pi_{t+1})-\rho)] - [E_t(y^n_{t+1}) - \frac{1}{\sigma}(r^n-\rho)] $

\begin{equation}
\tilde{y}_t = E_t(\tilde{y}_{t+1}) - \frac{1}{\sigma}(i_t-E_t(\pi_{t+1})-r_t^n)
\end{equation}

DIS equation equation determines the output gap given a path for the exogenous natural rate and the actual real rate. Then one can solve forward to yield: 
$$\tilde{y}_t = -\frac{1}{\sigma} E_t( \sum_{k=0}^\infty (r_{t+k}-r_{t+k}^n)$$
Output actually depends on expected policy rate (long-term interest rate). The output gap is proportional to the sum of current and anticipated deviations between the real interest rate and its natural counterpart. 


### Firms

Assume Cobb-Douglas technology $Y_{it}=A_tN_{it}^{1-\alpha}$. Each firm produces a differentiated good, but they all use an identical technology, where $A_t$ represents the level of technology, assumed to be common to all firms and to envolve exogenously over time. 

All firms face an identical isoelastic demand schedule, and  take the aggregate price level $P_t$ and aggregate consumption index $C_t$. 

They have monopolistic power, so can reset prices (monopolistic competition). Each firm may reset its price only with probability $(1-\theta)$ in any given period, independent of the time elapsed since the last adjustment. Each period a measure $(1-\theta)$ of producers reset their prices, while a fraction $\theta$ keep their prices unchanged ($\theta \rightarrow 1$, prices are fully fixed). 

The aggregate price dynamics can be approximated by a log-linear the aggregate price index around the steady state: $\pi_t = (1-\theta)(p_t^* - p_{t-1})$. 

### Optimal price setting
$$\max_{P_t^*} \sum_{k=0}^\infty \theta^k E_t\left\lbrace Q_{t,t+k} (P_t^* Y_{t+k|t} - \Psi_{t+k}(Y_{t+k|t})) \right\rbrace $$

s.t.
$$Y_{y+k|t}=\left(\frac{P_t^*}{P_{t+k}}\right)^{-\epsilon}C_{t+k}$$

where
- $P_t^*$: price that maximizes the current market value of the profits generated while that price remains effective. 
- $Q_{t,t+k} \equiv \beta^k (C_{t+k}/C_t)^\sigma(P_t/P_{t+k})$: stochastic discount factor for nominal payoffs.
- $\Psi()$: is the cost function.
- $Y_{t+k|t}$: output in the period t+k for a firm that the last reset its price in period t. 

Inflation is determined as follows: 
$$\pi_t=\beta E_t\left\lbrace \pi_{t+1}\right\rbrace + \lambda \hat{mc}_t$$
where 
- $\Theta \equiv \frac{1-\alpha}{1-\alpha+\alpha\epsilon} \leq 1$
- $\lambda \equiv \frac{(1-\theta)(1-\beta \theta)}{\theta}\Theta$
- $\hat{mc}_t$: deviation of the real marginal cost from steady state. 
if prices are sticky ($\theta$ close to 1) 


### New Keynesian Phillips Curve (NKPC)
$$\pi_t=\beta E_t\left\lbrace \pi_{t+1}\right\rbrace + \kappa \tilde{y}_t$$
where 
- $\kappa \equiv \lambda (\sigma+\frac{\varphi+\alpha}{1-\alpha})$
- if $\theta \rightarrow 0$, then $\kappa \rightarrow \infty$, so prices are fully flexible.
- if $\theta \rightarrow 1$, then $\kappa \rightarrow 0$, so prices are fully fixed

Iterating the NKPC forward
$$\pi_t=\sum_{i=0}^\infty \beta^i E_t \left\lbrace \kappa \tilde{y}_{t+i} \right\rbrace$$
Actual inflation depends on expected output gap. Output gap depends on expected policy. 

### Taylor rule

$$i_t=\rho + \phi_{\pi} \pi_t + \phi_{y}\tilde{y}_t+ \nu_t$$

## 2. Dynare model <a name="dynare_model"></a>

`Model Equations`
1. Phillips Curve: $\pi_t = \beta E_t \{ \pi_{t+1} \} + \kappa \tilde{y}_t+z_t$
2. IS-Equation: $\tilde{y}_t = -\frac{1}{\sigma}(i_t - E_t \{ \pi_{t+1} \} - r_t^n)  + E_t{\tilde{y}_{t+1}}$
3. Taylor rule: $i_t = \rho + \phi_{\pi} \pi_t + \phi_y \tilde{y}_t + \nu_t$
4. Natural rate: $r_t^n=\rho+\sigma \psi_{ya}^n E_t(a_{t+1}-a_t)$
5. Actual output: $y_t = a_t + (1-\alpha)n_t$
6. Natural output: $y_t^n= \psi_{ya}^n a_t + \vartheta_y^n $

7. Money Demand: $m_{r,t} = y_t-\eta i_t$ , where $m_{r,t}=m_t-p_t$
8. Technology shock: $a_t = \rho_a*a_{t-1} + \epsilon_t^a$
9. Monetary shock: $\nu_t = \rho_{\nu}*\nu_{t-1} + \epsilon_t^{\nu}$
10. Phillips curve shock: $z_t=\rho_{z}*z_{t-1} + \epsilon_t^{z}$

`Parameters` 
- $\beta = 0.99$
- $\sigma = 1$
- $\phi = 1$
- $\alpha = 1/3$
- $\epsilon = 6$
- $\eta = 4$
- $\theta = 2/3$
- $\phi_{\pi} = 1.5$
- $\phi_{y} = 0.5/4$
- $\psi_{ya}^n = (1+\phi)/(\sigma (1-\alpha)+\phi+\alpha)$
- $\Theta = (1-\alpha)/(1-\alpha+\alpha \epsilon)$
- $\lambda = (1-\theta) * (1 - \beta \theta)\Theta / \theta$
- $\kappa = \lambda (\sigma+(\phi+\alpha)/(1-\alpha))$
- $\rho = 1/ \beta -1$
- $\rho_{\nu} = 0.5$
- $\rho_{a} = 0.9$
- $\rho_{z}= 0.5$
